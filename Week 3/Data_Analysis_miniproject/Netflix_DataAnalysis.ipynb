# Display the number of movies by country using pie chart

```python
#  Filter out rows where 'type' is 'Movie'
movies_data = data[data['type'] == 'Movie']

# Count the number of movies by country
movies_by_country = movies_data['country'].value_counts()

# Plotting
plt.figure(figsize=(10, 8))
plt.pie(movies_by_country, labels=movies_by_country.index, autopct='%1.1f%%', startangle=140,labeldistance = 1.2)   # mentioning label distance as None to avoid overlapping of the countries names
plt.title('Distribution of Movies by Country on Netflix')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

# Counting the number of movies released in the specified year by the user

```python
# Filter out rows where 'type' is 'Movie'
movies_data = data[data['type'] == 'Movie']

# Get the unique release years of movies
unique_years = movies_data['release_year'].unique()

# Prompt user to input the year
user_year = int(input("Enter the year to find the number of movies released: "))

try:
    if user_year in unique_years:
        # Filter movies data for the specified year and count the number of movies
        movies_in_year = movies_data[movies_data['release_year'] == user_year]
        num_movies = len(movies_in_year)
        print(f"Number of movies released in {user_year}: {num_movies}")
    else:
        print("Invalid year! Please enter a valid release year.")
except ValueError:
    print("Invalid input! Please enter a valid year.")
```

# Giving the name of the movies released in the specified year by the user 

```python
movies_data = data[data['type'] == 'Movie']

# Get the unique release years of movies
unique_years = movies_data['release_year'].unique()

# Prompt user to input the year they want to search for
user_input = int(input("Enter a release year: "))

# Check if the users input is in the unique_years list
if user_input in unique_years:
    # Filter movies data for the specified year
    movies_in_year = movies_data[movies_data['release_year'] == user_input]
    # Get the names of the movies released in the specified year
    movie_names = movies_in_year['title']   #.tolist()
    print(f"Movies released in {user_input}:")
    for movie_name in movie_names:
        print(movie_name)
else:
    print("No movies found for the specified year.")
```

# Display the genre of the movies through histogram

```python
# Filter out rows where 'type' is 'Movie'
movies_data = data[data['type'] == 'Movie']

# Split genres column (assuming genres are separated by commas)
genres_split = movies_data['listed_in'].str.split(',').explode()

# Remove leading/trailing whitespaces
genres_split = genres_split.str.strip()

# Count the occurrence of each genre
genre_counts = genres_split.value_counts()

# Plotting
plt.figure(figsize=(12, 6))
genre_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Movie Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

# Displaying the rating of the movies versus the number of movies.

```python
# Filter out rows where 'type' is 'Movie'
movies_data = data[data['type'] == 'Movie']

# Split genres column (assuming genres are separated by commas)
rating_split = movies_data['rating'].str.split(',').explode()

# Remove leading/trailing whitespaces
rating_split = rating_split.str.strip()

# Count the occurrence of each genre
rating_counts = rating_split.value_counts()

# Plotting
plt.figure(figsize=(12, 6))
rating_counts.plot(kind='bar', color='skyblue')
plt.title('Rating of the movies')
plt.xlabel('Rating')
plt.ylabel('Number of Movies')
plt.xticks(rotation = 45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

# Displaying the genre of the tv show with the rating type 

```python

# Filter out rows where 'type' is 'Movie'
show_data = data[data['type'] == 'TV Show']

# Split genres column (assuming genres are separated by commas)
genres_split = show_data['listed_in'].str.split(',').explode()

# Remove leading/trailing whitespaces
genres_split = genres_split.str.strip()

# Count the occurrence of each genre
genre_counts = genres_split.value_counts()

# Plotting
plt.figure(figsize=(12, 6))

colors = plt.cm.tab10.colors[:len(genre_counts)]

for genre in genre_counts.index:
    # Filter movies by genre
    genre_movies = show_data[show_data['listed_in'].str.contains(genre)]
    # Count the occurrence of each rating within the genre
    rating_counts = genre_movies['rating'].value_counts().sort_index()
    # Plot the bar chart for the genre
    plt.bar(rating_counts.index, rating_counts.values, label=genre)

plt.title('Rating Distribution by Genre of TV Show')
plt.xlabel('Rating')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Trend analysis on the netflix dataset

```python
# Convert 'date_added' column to datetime
data['date_added'] = pd.to_datetime(data['date_added'])

# Extract year from 'date_added' column
data['year_added'] = data['date_added'].dt.year

# Count the number of titles added each year
titles_added_by_year = data['year_added'].value_counts().sort_index()

# Plotting the trend analysis using a line graph
plt.figure(figsize=(12, 6))
plt.plot(titles_added_by_year.index, titles_added_by_year.values, marker='o')

plt.title('Trend Analysis of Titles Added to Netflix')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.xticks(range(min(data['year_added']), max(data['year_added']) + 1), rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter out rows where 'type' is 'Movie'
movie_data = data[data['type'] == 'Movie']

# Split genres column (assuming genres are separated by commas)
genres_split = movie_data['listed_in'].str.split(',').explode()

# Remove leading/trailing whitespaces
genres_split = genres_split.str.strip()

# Count the occurrence of each genre
genre_counts = genres_split.value_counts()

# Plotting
plt.figure(figsize=(12, 6))

# Create scatter plot
plt.scatter(genre_counts.values, genre_counts.index, color='skyblue', alpha=0.6)

plt.title('Movie Distribution by Genre')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Filter the dataset for movies
movie = data[data['type'] == 'Movie']

# Apply the 'fivethirtyeight' style
plt.style.use('fivethirtyeight')

# Create a color palette for each unique genre
palette = sns.color_palette("husl", len(movie['listed_in'].unique()))

# Create scatter plot with different colors for each genre
plt.figure(figsize=(12, 8))  # Increase the figure size here
for i, genre in enumerate(movie['listed_in'].unique()):
    plt.scatter(movie['release_year'][movie['listed_in'] == genre], 
                movie['duration'][movie['listed_in'] == genre], 
                label=genre, 
                color=palette[i])

# Add labels and title
plt.xlabel('Release Year')
plt.ylabel('Duration')
plt.title('Scatter Plot of Movie Duration by Genre')
plt.xticks(rotation=45)

# Position the legend outside the plot area with reduced font size and improved layout
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small', ncol=2)

# Show plot
plt.tight_layout()  # Adjust layout to prevent clipping of legend
plt.show()
```
